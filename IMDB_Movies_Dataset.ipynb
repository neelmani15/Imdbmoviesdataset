```python
import pandas as pd
import os
import numpy as np
from jupyter_dash import JupyterDash
```

```python
from bokeh.io import curdoc,output_notebook
from bokeh.layouts import column,layout
from bokeh.models import CustomJS,ColumnDataSource,Slider,TextInput,Select,Div
from bokeh.plotting import figure
output_notebook()
```

```python
df = pd.read_csv("./Movies_dataset/IMDb movies.csv")
df.head()
```

```python
div=Div(text="""Here I am presenting the data of <b>IMDB MOVIES Dataset</b> is taken from Kaggle.I have presented through Bokeh and the Link of data is given <a href="https://www.kaggle.com/stefanoleone992/imdb-extensive-dataset">HERE</a>.""")
```

```python
df['reviews_from_users']=df['reviews_from_users'].fillna(0)
df['reviews_from_critics']=df['reviews_from_critics'].fillna(0)
df['review']=df['reviews_from_users']+df['reviews_from_critics']
df.head()
```

```python
df['color'] = np.where(df['review']>=15,'orange','grey')
df.head()
```

```python
df1=df[df['year'].str.contains('TV Movie 2019').fillna(False)]
```

```python
i=df1.index
i
```

```python
df = df.drop(i,axis=0)
```

```python
df['year']=pd.to_numeric(df['year'])
```

```python
df.year.astype(int)
```

```python
axis_map={
    "Reviews from user":"reviews_from_users",
    "Reviews from critics":"reviews_from_critics",
    "Year":"year",
    "Rating":"avg_vote",
    "Date Published":"date_published",
    "Review":"review"
}
```

```python
year = Slider(title="Year of released",value=1950,start=1890,end=2020,step=1)
genre = TextInput(title="Genre")
rating = Slider(title="Rating",value=3,start=0,end=10,step=0.5)
duration = Slider(title="Length of Movies",value=100,start=40,end=808,step=10)
language = TextInput(title="Language")
director = TextInput(title="Name of Director")
writer = TextInput(title="Name of writer")
country = TextInput(title="Movie launched country")
actor = TextInput(title="Actor Name")
x_axis = Select(title="X Axis", options=sorted(axis_map.keys()), value="Year")
y_axis = Select(title="Y Axis", options=sorted(axis_map.keys()), value="Rating")
```

```python
source = ColumnDataSource(data=dict(x=[],y=[],title=[],year=[],duration=[],color=[]))
```

```python
TOOL=[
    ("Title","@title"),
    ("Year","@year"),
    ("Length","@duration")
]
```

```python
p=figure(plot_height=600,plot_width=700,toolbar_location='above',tooltips=TOOL)
p.circle(x="x",y="y",source=source,size=5,color='color')
```

```python
def movies():
    ge=genre.value.strip()
    lan=language.value.strip()
    di=director.value.strip()
    wri=writer.value.strip()
    coun=country.value.strip()
    ac=actor.value.strip()
    selects=df[
        (df.avg_vote >= rating.value) &
        (df.year >= year.value) &
        (df.duration >= duration.value)
    ]
    if(genre !=""):
        selects=selects[selects.genre.str.contains(ge)==True]
        selects=selects[selects.language.str.contains(lan)==True]
        selects=selects[selects.director.str.contains(di)==True]
        selects=selects[selects.country.str.contains(coun)==True]
        selects=selects[selects.actors.str.contains(ac)==True]
        selects=selects[selects.writer.str.contains(wri)==True]
    return selects
```

```python
def update():
    df1=movies()
    x_name=axis_map[x_axis.value]
    y_name=axis_map[y_axis.value]
    
    p.xaxis.axis_label=x_axis.value
    p.yaxis.axis_label=y_axis.value
    p.title.text = "%d movies selected" % len(df1)
    source.data=dict(
        x=df1[x_name],
        y=df1[y_name],
        title=df1['title'],
        year=df1['year'],
        color=df1['color'],
        duration=df1['duration']
    )
```

```python
controls=[year,rating,genre,duration,country,language,director,writer,actor,x_axis,y_axis]
for control in controls:
    control.on_change('value',lambda attr,old,new:update())
```

```python
m=column(*controls,width=300,height=1000)
m.sizing_mode="fixed"
l=layout([
    [div],
    [m,p],
],sizing_mode="scale_both")
curdoc().add_root(l)
curdoc.title="Movies"
update()
```
